## Part 1

```python
import numpy as np
```

```python
class Hand(object):
    def __init__(self,hand,bid):
        self.order_type = ["Five","Four","Full","Three","Two Pair","One Pair","High Card"]
        self.order_card = ["A","K","Q","J","T","9","8","7","6","5","4","3","2"]
        self.hand = hand
        self.bid = bid
        self.type = self.get_type()
    
    def get_type(self):
        values,count = np.unique(self.hand, return_counts=True)
        if max(count) == 5:
            return "Five"
        elif max(count) == 4:
            return "Four"
        elif max(count) == 3:
            if 2 in count:
                return "Full"
            else:
                return "Three"
        elif max(count) == 2:
            if sum(count==2) == 2:
                return "Two Pair"
            else:
                return "One Pair"
        else:
            return "High Card"
    
    def __eq__(self, other):
        if isinstance(other, Hand):
            return self.hand == other.hand
        else:
            return False
    
    def __ne__(self,other):
        if isinstance(other, Hand):
            return self.hand != other.hand
        else:
            return True
    
    def __lt__(self,other):
        if isinstance(other, Hand):
            if self.order_type.index(self.type) < self.order_type.index(other.type):
                return False
            elif self.order_type.index(self.type) > self.order_type.index(other.type):
                return True
            else:
                for i in range(len(self.hand)):
                    if self.order_card.index(self.hand[i]) < self.order_card.index(other.hand[i]):
                        return False
                    elif self.order_card.index(self.hand[i]) > self.order_card.index(other.hand[i]):
                        return True
                return False

        
    def __gt__(self,other):
        if isinstance(other, Hand):
            if self.order_type.index(self.type) < self.order_type.index(other.type):
                return True
            elif self.order_type.index(self.type) > self.order_type.index(other.type):
                return False
            else:
                for i in range(len(self.hand)):
                    if self.order_card.index(self.hand[i]) < self.order_card.index(other.hand[i]):
                        return True
                    elif self.order_card.index(self.hand[i]) > self.order_card.index(other.hand[i]):
                        return False
                    return False
        
    def __le__(self,other):
        if isinstance(other, Hand):
            if self.order_type.index(self.type) < self.order_type.index(other.type):
                return False
            elif self.order_type.index(self.type) > self.order_type.index(other.type):
                return True
            else:
                for i in range(len(self.hand)):
                    if self.order_card.index(self.hand[i]) < self.order_card.index(other.hand[i]):
                        return False
                    elif self.order_card.index(self.hand[i]) > self.order_card.index(other.hand[i]):
                        return True
                return True
        
    def __ge__(self,other):
        if isinstance(other, Hand):
            if self.order_type.index(self.type) < self.order_type.index(other.type):
                return False
            elif self.order_type.index(self.type) > self.order_type.index(other.type):
                return True
            else:
                for i in range(len(self.hand)):
                    if self.order_card.index(self.hand[i]) < self.order_card.index(other.hand[i]):
                        return False
                    elif self.order_card.index(self.hand[i]) > self.order_card.index(other.hand[i]):
                        return False
                return True
    
    def __str__ (self):
        return f"{self.hand},{self.type},{self.bid}"
```

```python
hands = []
with open("input") as input_file:
    for l in input_file.readlines():
        hand,bid = l.split(" ")
        hand = [c for c in hand]
        bid = int(bid)
        h = Hand(hand,bid)
        hands.append(h)
```

```python
hands.sort()
ret = 0
for i,h in enumerate(hands,1):
    ret += i*h.bid
ret
```
